```python
import pandas as pd

#pegando base de dados .csv
df = pd.read_csv('vendas_acai.csv')
df = df.dropna()

#mudando para data
df['data_venda'] = pd.to_datetime(df['data_venda'])
df['mensal'] = df['data_venda'].dt.month

#filtros
produto_mais_vendido = df.groupby('produto')['total_venda'].sum().sort_values(ascending=False).reset_index()
valores = df.groupby('produto')['total_venda'].sum().sort_values(ascending=False)
lucrativo = produto_mais_vendido.loc[0, 'produto']
top_5_produtos = produto_mais_vendido.head(5)
top_5_valores = valores.head(5)
mensal_vendas = df.groupby('mensal')['total_venda'].sum().sort_values(ascending=False).reset_index()
mmvr = mensal_vendas.head(5)

#test1
display(df)
print('O produto mais lucrativo é:', lucrativo)

#criando novo dicionario
relatorio = {
    'Produto mais Vendido': top_5_produtos['produto'].values.tolist(),
    'N. do Mês': mmvr['mensal'].values.tolist(),
    'Venda Total': top_5_produtos['total_venda'].values.tolist()
}
#transformando o dicionario para DataFrame
df2R = pd.DataFrame(relatorio)
#mudando o DataFrame para relatorio.csv
df2R.to_csv('Relatorio.csv', index=False)
```
